```python
import pandas as pd
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
import os
```

```python

# Carica il modello ResNet50 pre-addestrato su Stanford Dogs Dataset
model = ResNet50(weights='imagenet')

# Percorso del dataset
dataset_path = 'percorso/del/tuo/dataset'

# Carica il mapping delle classi (razze di cani)
class_mapping = pd.read_csv(os.path.join(dataset_path, 'labels.csv'))

def load_and_preprocess_image(image_path):
    # Carica l'immagine e ridimensiona alle dimensioni richieste dal modello ResNet50
    img = image.load_img(image_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    return img_array

def predict_dog_breed(image_path, model, class_mapping):
    # Carica e prepara l'immagine
    img_array = load_and_preprocess_image(image_path)

    # Ottieni le previsioni
    predictions = model.predict(img_array)

    # Decodifica le previsioni
    decoded_predictions = decode_predictions(predictions)

    # Estrai la razza del cane dalla previsione principale
    predicted_class = decoded_predictions[0][0][0]

    # Trova il nome della razza nella tabella di mapping delle classi
    dog_breed = class_mapping[class_mapping['breed_id'] == predicted_class]['breed'].values[0]

    return dog_breed

# Percorso dell'immagine del cane
image_path = 'percorso/della/tua/foto_di_cane.jpg'

# Ottieni la razza del cane
predicted_breed = predict_dog_breed(image_path, model, class_mapping)

# Stampa la razza predetta
print(f"La razza del cane è: {predicted_breed}")
```
